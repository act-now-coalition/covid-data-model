```python
import os
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from pyseir import load_data

# Global settings for the notebook.
base_path_new = '/Users/natashawoods/Desktop/later.nosync/covid_act_now.nosync/covid-data-model/output/web_ui/state/AL.0.json'
comparison_cols = [(8, 'current_icu'), (10, 'all_infected'), (11, 'dead'), (12, 'beds')]
```

```python
df_new = pd.read_json(os.path.join(base_path_new), convert_dates=[1]) 
```

```python
plt.figure(figsize=(16, 16))

for i_plt, (col, compartment) in enumerate(comparison_cols):
    plt.subplot(2, 2, i_plt + 1)
    plt.plot(df_new[1] - datetime.timedelta(days=1), df_new[col], label=compartment)
    plt.legend()
    plt.xlim(datetime.datetime.strptime('3/15/2020', '%m/%d/%Y'), datetime.datetime.strptime('7/15/2020', '%m/%d/%Y'))
    _ = plt.xticks(rotation=30)
```
